# Kom igang med Python

## Get python running

Den här guiden visar steg för steg hur du installerar en minimal Python-miljö på Windows för att kunna köra koden i detta projekt, även om du aldrig installerat Python tidigare.

## 1. Installera Python

1. Öppna https://www.python.org/downloads/.
2. Ladda ner senaste stabila Python 3-versionen (t.ex. 3.12 eller nyare).
3. Kör installationsfilen.
4. Viktigt: markera kryssrutan **Add python.exe to PATH** innan du klickar på Install.
5. Starta om terminalen när installationen är klar.

## 2. Kontrollera att Python fungerar

Öppna **PowerShell** och kör:

```powershell
python --version
pip --version
```

Om du får ett versionsnummer är installationen klar.

## 3. Skapa en virtuell miljö i projektmappen

I projektets rotmapp (den som innehåller mappar som `notebooks`, `data` och `raw scripts`) kör du:

```powershell
python -m venv .venv
.\.venv\Scripts\Activate.ps1
```

Tips: Om PowerShell blockerar aktivering, kör en gång:

```powershell
Set-ExecutionPolicy -Scope CurrentUser RemoteSigned
```

## 4. Installera minimala paket för detta projekt

Med den virtuella miljön aktiv kör du:

```powershell
python -m pip install --upgrade pip
pip install jupyterlab notebook ipykernel pandas numpy scipy matplotlib seaborn pastas
```

Detta räcker för att köra den vanligaste analyskoden i mapparna `data`, `notebooks_temp` och `raw scripts`.

## 5. Valfria extrapaket (bara om du behöver dem)

Vissa script använder fler bibliotek. Installera bara vid behov:

```powershell
pip install numba pillow adjustText pyeto
```

För kartskript kan även `cartopy` behövas, men det kan kräva extra systemberoenden och är inte del av den minimala starten.

## 6. Starta Jupyter och välj miljö

Kör i samma aktiverade terminal:

```powershell
python -m ipykernel install --user --name exempelforlopp --display-name "Python (.venv) exempelforlopp"
jupyter lab
```

När Jupyter öppnas: välj kerneln **Python (.venv) exempelforlopp** i notebooken.

````python
import sys
import pandas as pd
import numpy as np
import scipy
import matplotlib

print("Python:", sys.version.split()[0])
print("pandas:", pd.__version__)
print("numpy:", np.__version__)
print("scipy:", scipy.__version__)
print("matplotlib:", matplotlib.__version__)
````

## 7. Vanliga fel och lösningar

- **`python` hittas inte**: starta om PowerShell efter installation eller installera om Python med PATH ikryssat.
- **Fel miljö i notebook**: byt kernel till din `.venv`.
- **Modul saknas**: installera paketet i samma aktiverade miljö och starta om kernel.
